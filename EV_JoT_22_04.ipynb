**Introduction**

The concept of EV was not popular when Tesla released its first electric car in 2008. But today, the customer perception has changed and every major vehicle manufacturer has now entered into this market. Despite the Covid 19 pandemic, EV sales surged in the later half of [2020](https://www.iea.org/reports/global-ev-outlook-2021/policies-to-promote-electric-vehicle-deployment) with countries around the world offering [subsidies and incentives](https://www.weforum.org/agenda/2021/06/electric-vehicles-financial-incentives/) on purchase of these vehicles. Governments around the globe are already talking about banning sale of petrol and diesel cars, which will further drive EV sales. As per a report from [Mckinsey](https://www.mckinsey.com/industries/automotive-and-assembly/our-insights/why-the-automotive-future-is-electric), Europe will be a global leader in EV sales with 60 percent sales by 2030.  Apart from being stylish and appealing, they are better for the [environment](https://www.edfenergy.com/for-home/energywise/electric-cars-and-environment),  and emit fewer greenhouse gases and air pollutants than petrol or diesel cars.

As more and more models are being launched, it is essential to choose carefully which EV to buy. Earlier, the decision was based mostly on the car's range. But today, with a lot of options available, it best to consider all features and see which one fits our lifestyle best. The reason we have choosen this data is .... rising prices of petrol, so we are looking at cost effective alternatives for cars. 

Goal of this project is to analyse the performance of different EVs available in market by looking at below parameters:

**WHERE AND HOW TO CHARGE THE VEHICLE: WHAT IS THE BATTERY SIZE? DOES THE CAR HAVE RAPID CHARGING?** (The cost to charge an EV depends on the size of the battery and where its being charged. If the car will be used for long trips, it will be best if it has rapid charging. Can also check how quickly it can charge at a rapid charging point. While many cars regain 80% of their range in 30 minutes, some take longer.)

**CARS OPERATING RANGE: WHICH CARS HAVE THE LONGEST RANGE?** (Which EV can do the longest number of miles per charge? Depending on how far a customer has to drive or if they regularly take long trips, its worth considering a model with a longer range.)

**COMFORT: DOES THE VEHICLE MEET ME AND MY FAMILY NEEDS?** (Does the car have sufficient room for a family or not.)

**EFFICIENCY AND PERFORMANCE:** (EV's produce maximum torque isntantly from standstill, so acceleration can be measured to show their efficiency) acceleration, engine power, torque, battery size, top speed

# **DATA PREPARATION**

```python
import pandas as pd
import numpy as np
```

**Load the first data set**

```python
url1 = 'https://github.com/JoLRThomson/CMM202JoT/blob/main/ElectricCarData_Norm.csv?raw=1'
EV_data = pd.read_csv(url1)
EV_data
```

**Load the 2nd data set**

```python
url2 = 'https://github.com/JoLRThomson/CMM202JoT/blob/main/FEV-data.csv?raw=1'
FEV_data = pd.read_csv(url2, encoding = 'latin1')
FEV_data
```

**Clean the first data set**

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

# functions to clean the data
def clean_accel(col):
    return col.strip('sec ')

def clean_speed(col):
    return col.strip('km/h ')

def clean_range(col):
    return col.strip('km ')

def clean_efficiency(col):
    return col.strip('Wh/km ')

def clean_fast_charge(col):
    return col.strip('km/h ')
       
def clean_power_train(col):
    return col.replace(' Wheel Drive', '')

# apply functions to columns
EV_data['Accel'] = EV_data['Accel'].apply(clean_accel)
EV_data['TopSpeed'] = EV_data['TopSpeed'].apply(clean_speed)
EV_data['Range'] = EV_data['Range'].apply(clean_range)
EV_data['Efficiency'] = EV_data['Efficiency'].apply(clean_efficiency)
EV_data['FastCharge'] = EV_data['FastCharge'].apply(clean_fast_charge)
EV_data['PowerTrain'] = EV_data['PowerTrain'].apply(clean_power_train)

# mapping for yes and no values
mapping = {'Rapid charging possible':'Yes', 'Rapid charging not possible':'No'}
EV_data['RapidCharge'] = EV_data['RapidCharge'].map(mapping)

# deal with missing values
EV_data['FastCharge'] = EV_data['FastCharge'].replace('-', np.nan)

# create a new column called car full name which is make + model to allow merging of both datasets
EV_data['Car full name'] = EV_data['Brand']+EV_data['Model']

# create a new column for segment description 
seg_desc = {'A':'Mini', 'B':'Supermini', 'C':'Small Family', 'D':'Large Family', 'E':'Executive', 'F':'Luxury', 'S':'Sport', 'N':'Commercial'}
EV_data['Segment Description'] = EV_data['Segment'].map(seg_desc)

# remove unwanted columns
EV_data = EV_data.drop(labels=['Model'], axis=1)

# rename columns
EV_data = EV_data.rename(columns= {'Brand':'Make', 'Accel':'Acceleration', 'TopSpeed':'Maximum speed', 'Range':'Range', 'Efficiency':'Battery efficiency', 'FastCharge':'Rate of charge', 'RapidCharge':'Rapid Charge Possible?'})

EV_data.head()
```

**Remove whitespace from car names to allow merge on car names**

```python
EV_data['Car full name'] = EV_data['Car full name'].str.strip()
```

**Set 'Car full name' to be the index**

```python
EV_data = EV_data.set_index('Car full name', drop=True)
EV_data = EV_data.sort_index()
```

```python
EV_data.index.is_unique # this shows there are duplicate car names in the index
```

**Remove unwanted columns from FEV_data**

```python
FEV_data = FEV_data.drop(labels=['Model', 'Drive type', 'Range (WLTP) [km]','Number of seats', 'Number of doors'], axis=1)
# set index to car full name sorted a to z and convert to string
FEV_data = FEV_data.set_index('Car full name', drop=True)
FEV_data.index = FEV_data.index.map(str)
FEV_data = FEV_data.sort_index(ascending=True)
FEV_data.head()
```

**Rename index values to ensure consistency in car name across both sets of data**

```python
FEV_data = FEV_data.rename(index={'Volkswagen e-up!':'Volkswagen e-Up!', 
                                  'Volkswagen ID.4 1st':'Volkswagen ID.4', 
                                  'Tesla Model X Long Range Plus':'Tesla Model X Long Range',
                                  'Tesla Model S Long Range Plus':'Tesla Model S Long Range',
                                  'Tesla Model 3 Performance':'Tesla Model 3 Long Range Performance',
                                  'Tesla Model 3 Long Range':'Tesla Model 3 Long Range Dual Motor',
                                  'Smart fortwo EQ':'Smart EQ fortwo coupe',
                                  'Smart forfour EQ':'Smart EQ forfour',
                                  'Skoda Citigo-e iV':'Skoda CITIGOe iV',
                                  'Renault Zoe R135':'Renault Zoe ZE50 R135',
                                  'Renault Zoe R110':'Renault Zoe ZE50 R110',
                                  'Porsche Taycan 4S (Performance)':'Porsche Taycan 4S',
                                  'Porsche Taycan 4S (Performance Plus)':'Porsche Taycan 4S Plus',
                                  'Peugeot e-2008':'Peugeot e-2008 SUV',
                                  'Nissan e-NV200 evalia':'Nissan e-NV200 Evalia',
                                  'Mercedes-Benz EQV (long)':'Mercedes EQV 300 Long',
                                  'Mercedes-Benz EQC':'Mercedes EQC 400 4MATIC',
                                  'Kia e-Soul 64kWh':'Kia e-Soul 64 kWh',
                                  'Kia e-Soul 39.2kWh':'Kia e-Soul 39 kWh',
                                  'Kia e-Niro 64kWh':'Kia e-Niro 64 kWh',
                                  'Kia e-Niro 39.2kWh':'Kia e-Niro 39 kWh',
                                  'Hyundai Kona electric 64kWh':'Hyundai Kona Electric 64 kWh',
                                  'Hyundai Kona electric 39.2kWh':'Hyundai Kona Electric 39 kWh',
                                  'Hyundai Ioniq electric':'Hyundai IONIQ Electric',
                                  'DS DS3 Crossback e-tense':'DS 3 Crossback E-Tense',
                                  'Citron -C4':'Citroen e-C4',
                                  'Citron -Spacetourer (M)': 'Citroen e-Spacetourer (M)',
                                  'BMW i3s':'BMW i3s 120 Ah',
                                  'BMW i3':'BMW i3 120 Ah',
                                  'Audi e-tron Sportback S quattro':'Audi e-tron S Sportback 55 quattro',
                                  'Audi e-tron S quattro':'Audi e-tron S 55 quattro'})
```

**Merge both sets of data keeping all the data from EV_data**

```python
new_merge = new_merge.astype({"Acceleration":"float"})
new_merge = new_merge.astype({"Battery efficiency":"int"})
new_merge = new_merge.astype({"Rate of charge":"float"})
new_merge = new_merge.astype({"Maximum speed":"int"})
new_merge = new_merge.astype({"Range": "int"})
print(new_merge.dtypes)
```
