# **DATA PREPARATION**

```python
import pandas as pd
import numpy as np
```

Load the first data set

```python
url1 = 'https://github.com/JoLRThomson/CMM202JoT/blob/main/ElectricCarData_Norm.csv?raw=1'
EV_data = pd.read_csv(url1)
EV_data.head()
```

Load the 2nd data set

```python
url2 = 'https://github.com/JoLRThomson/CMM202JoT/blob/main/FEV-data.csv?raw=1'
FEV_data = pd.read_csv(url2, encoding = 'latin1')
FEV_data.head()
```

Cleaning the first data set

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

# functions to clean the data
def clean_accel(col):
    return col.strip('sec ')

def clean_speed(col):
    return col.strip('km/h ')

def clean_range(col):
    return col.strip('km ')

def clean_efficiency(col):
    return col.strip('Wh/km ')

def clean_fast_charge(col):
    return col.strip('km/h ')
       
def clean_power_train(col):
    return col.replace(' Wheel Drive', '')

# apply functions to columns
EV_data['Accel'] = EV_data['Accel'].apply(clean_accel)
EV_data['TopSpeed'] = EV_data['TopSpeed'].apply(clean_speed)
EV_data['Range'] = EV_data['Range'].apply(clean_range)
EV_data['Efficiency'] = EV_data['Efficiency'].apply(clean_efficiency)
EV_data['FastCharge'] = EV_data['FastCharge'].apply(clean_fast_charge)
EV_data['PowerTrain'] = EV_data['PowerTrain'].apply(clean_power_train)

# mapping for yes and no values
mapping = {'Rapid charging possible':'Yes', 'Rapid charging not possible':'No'}
EV_data['RapidCharge'] = EV_data['RapidCharge'].map(mapping)

# deal with missing values
EV_data['FastCharge'] = EV_data['FastCharge'].replace('-', np.nan)

# create a new column called car full name which is make + model to allow merging of both datasets
EV_data['Car full name'] = EV_data['Brand']+EV_data['Model']

# create a new column for segment description 
seg_desc = {'A':'Mini', 'B':'Supermini', 'C':'Small Family', 'D':'Large Family', 'E':'Executive', 'F':'Luxury', 'S':'Sport', 'N':'Commercial'}
EV_data['Segment Description'] = EV_data['Segment'].map(seg_desc)

# remove unwanted columns
EV_data = EV_data.drop(labels=['Model'], axis=1)

# rename columns
EV_data = EV_data.rename(columns= {'Accel':'Acceleration 0-100 kph[s]', 'TopSpeed':'Maximum speed [kph]', 'Range':'Range (WLTP) [km]', 'Efficiency':'Battery efficiency [Wh/km]', 'FastCharge':'Rate of charge [km/h]', 'RapidCharge':'Rapid Charge Possible?'})

EV_data.head()
```

# **DATA VISUALIZATION**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

There are some improtant things to consider before deciding to buy an electric car. Factors like the car's range, charging options, a family's needs and budget need to be evaluated properly. Through data visualization, we would analyse the data that we have and answer these questions. 

a) **How far does one need to travel?**

Range is an important factor to consider before choosing an elelctric car. EVs offer a shorter range as compared to a diesel or petrol cars, so the range of the EV should be more than the average distance travelled by the user everyday.  

```python
px.scatter(EV_Merged,
          x='Car Brand',
          y='Range (WLTP) [km]',
          size='Range (WLTP) [km]', 
          color='Car Brand', 
          height = 800,
          title='Car Range')
```

There are many brands which have cars with small to medium range. Tesla has models with the highest range among all the brands.  If the car will be used for commuting to office and general runs around the city, one doesn't need a very high range. Whereas if it will be used to travel long distances or regular trips out of the city, a higher range car would be needed.  Depending on individual travel length requirements, a suitable brand can be chosen.

b) **What are the charging options in the car?**

Electric car charges come in different speeds : slow, fast and rapid. Many cars come with a charging cable that can be plugged into a normal socket at a house. But this is slow and depending on the car's range and size of the battery, it can take many hours and a fast charger can usually charge a car overnight. Whereas rapid charges are able to charge a car very fast and these charging stations can be found at many public places today. So, depending on the requirement, it is important to check if the car has rapid charging or not.

```python
px.bar(EV_Merged, 
       x='Car Brand', 
       y='PriceEuro',
       color='Rapid Charge Possible?',
       hover_data=['Range (WLTP) [km]'],
       title='Is Rapid charge possible?')
```

It can be seen that all the brands except "Smart" have cars in which rapid charge is possible. This shows that all the brands have equipped their models with rapid charge option as this gives the user an ease of operation.  

Apart from rapid charging, other factors to consider are the car's charging capacity and type of connector it has.  

```python
px.bar(EV_Merged, 
       x='Car Brand', 
       y='Maximum charging power',
       color='Maximum charging power', 
       title='DC Charging Power (kW)')
```

The above graph analyzes charging capcity of the vehicles, and it can be seen that Porsche make the fastest charging cars at 270kW, followed by Tesla and Audi cars with charging speed of 150kW.

```python
px.pie(EV_Merged, names='PlugType',  
       title='Charging Connector Type')
```

Type 2 CCS is the most common type of connector port installed in the cars.  

c) **What will be the car used for?**

It will be important to consider if the car suits its owner's personal preference or family needs. Some factors to evaluate are size of the car and number of seats.

e) **Is there a correlation between certain features?**

There are certain other features, like acceleration and speed, that are important when considering an EV. A Heatmap can be plotted to check if there is a correlation between these features.

```python
cars_corr = EV_Merged[['Maximum charging power', 'Range (WLTP) [km]', 'Battery efficiency [Wh/km]', 'Maximum speed [kph]_x', 'Rate of charge [km/h]', 'Maximum torque [Nm]', 'Battery capacity [kWh]', 'Engine power [KM]']]
cars_corr.head()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap((cars_corr).corr(), annot=True, cmap='coolwarm')
```

It can be seen that there is a strong coorelation between Engine power, Maximum speed, Torque and Battery capacity. Battery efficiency and range are correlated with battery capacity. 

EV's produce maximum torque instantly from standstill, so acceleration can be mapped with maximum speed to analyze if they are correlated.
